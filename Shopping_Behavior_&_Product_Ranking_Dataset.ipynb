```python
# Create a summary table for the target variable

# Count values
counts = df["Frequency of Purchases"].value_counts()

# Proportion (percentage)
proportions = df["Frequency of Purchases"].value_counts(normalize=True)

# Combine into one DataFrame
target_summary = pd.DataFrame({
    "Value": counts.index,
    "Count": counts.values,
    "Proportion": proportions.values
})

# Reset index for clean table
target_summary.reset_index(drop=True, inplace=True)

target_summary
```

Comment:
During the initial inspection of the target variable Frequency of Purchases, two semantic inconsistencies were identified.
The dataset contains pairs of categories that describe the same purchasing frequency but are labeled differently:

"Bi-Weekly" and "Fortnightly" both mean once every two weeks.

"Every 3 Months" and "Quarterly" both mean once every three months.

To ensure consistency and avoid misleading class separation during EDA and modelling, these categories were merged as follows:

Bi-Weekly and Fortnightly to Fortnightly

Every 3 Months and Quarterly to Quarterly

This clean version of the target variable will be used for all further EDA and modelling steps.



---

Since the inconsistencies in the target variable require data cleaning before any meaningful EDA can be performed, a new DataFrame will be created as a working copy. This separate DataFrame will be used for all cleaning steps to ensure that the original dataset remains intact. All subsequent transformations, including the correction of duplicated frequency categories, will be applied only within this new DataFrame.

```python
# Create a deep copy of the original DataFrame for cleaning
df_cleaned = df.copy(deep=True)
```

```python
# Replace duplicate frequency labels in the cleaned DataFrame

df_cleaned["Frequency of Purchases"] = df_cleaned["Frequency of Purchases"].replace({
    "Bi-Weekly": "Fortnightly",
    "Every 3 Months": "Quarterly"
})
```

```python
# Recalculate the summary table for the cleaned target variable

counts = df_cleaned["Frequency of Purchases"].value_counts()
proportions = df_cleaned["Frequency of Purchases"].value_counts(normalize=True)

target_summary_cleaned = pd.DataFrame({
    "Value": counts.index,
    "Count": counts.values,
    "Proportion": proportions.values
}).reset_index(drop=True)

target_summary_cleaned
```

```python
# Create an ordinal mapping for the cleaned target categories
ordinal_map = {
    "Weekly": "1_Weekly",
    "Fortnightly": "2_Fortnightly",
    "Monthly": "3_Monthly",
    "Quarterly": "4_Quarterly",
    "Annually": "5_Annually"
}

# Apply the mapping to the cleaned DataFrame
df_cleaned["Frequency of Purchases"] = df_cleaned["Frequency of Purchases"].map(ordinal_map)

# Recalculate the summary table after renaming
counts = df_cleaned["Frequency of Purchases"].value_counts()
proportions = df_cleaned["Frequency of Purchases"].value_counts(normalize=True)

target_summary_ordered = pd.DataFrame({
    "Value": counts.index,
    "Count": counts.values,
    "Proportion": proportions.values
})

# Sort rows alphabetically by the new Value column (which equals the correct order)
target_summary_ordered = target_summary_ordered.sort_values(by="Value").reset_index(drop=True)

target_summary_ordered
```

The target variable contained logically duplicated categories, which made the original set of seven classes inconsistent and unsuitable for reliable EDA or modelling. After identifying these inconsistencies, the target values were cleaned by merging equivalent categories and then converting the final five categories into an explicitly ordered format. Each category was assigned a numerical prefix reflecting its true ordinal position (from most frequent purchases to least frequent). This prefix ensures correct alphabetical sorting and preserves the ordinal nature of the target for all subsequent analysis steps.

```python
import matplotlib.pyplot as plt

# Plot distribution of the target variable (horizontal bar chart)
plt.figure(figsize=(8, 5))

df_cleaned["Frequency of Purchases"].value_counts().sort_index().plot(
    kind="barh"
)

plt.title("Frequency of Purchases Distribution")
plt.xlabel("Count")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```

Target Variable Distribution: Frequency of Purchases

This chart provides a visual overview of the cleaned and ordered target variable. After resolving duplicated category labels and converting the values into an ordinal format, the distribution shows the relative frequency of each purchasing pattern. The categories range from the most frequent buyers (â€œ1_Weeklyâ€) to the least frequent buyers (â€œ5_Annuallyâ€).

The bar lengths confirm that the classes remain well balanced even after combining redundant categories, which is ideal for a multi-class classification problem. The orderly prefixes ensure correct alphabetical sorting and make the ordinal structure visually clear for subsequent analysis and modeling.
